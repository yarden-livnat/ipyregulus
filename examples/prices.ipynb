```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
current = 750
strike = 1000
premium = 96.65
```

```python
prices = np.array(range(500, 2000))
```

```python
def call(v, strike, premium):
    if v < strike:
        return -premium
    return v-strike-premium
```

```python
p = pd.Series([call(v, strike, premium) for v in prices],  index=range(500, 2000))
```

```python
p.plot()
```

```python
df1 = pd.DataFrame({'premium': [premium for v in prices],
                    'stock': [v*premium/current for v in prices], 
                    'option': [call(v, strike, premium) for v in prices]
                   },
                   index=prices)
```

```python
df1.plot()
```
